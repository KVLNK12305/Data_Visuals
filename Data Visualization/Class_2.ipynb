```python
import matplotlib.pyplot as plt
import pandas as pd

# using the iris dataset
url = 'https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv'
iris = pd.read_csv(url)
iris.head()

#basic line plot
plt.plot(iris['sepal_length'][:30],marker = 'o',ls='--',color='r')
plt.title("Sepal Length Analysis")
plt.ylabel("Sepal Length")
plt.xlabel("Sample Index")
plt.grid()
plt.legend(['Sepal Length'],loc='upper right')

```

```python
import matplotlib.pyplot as plt
import pandas as pd

# using the iris dataset
url = 'https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv'
iris = pd.read_csv(url)
iris.head()

#basic line plot
plt.plot(iris['sepal_length'][:30],marker = 'o',ls='--',color='b')
plt.plot(iris['sepal_width'][:30],marker = 'o',ls='--',color='r')
plt.plot(iris['petal_length'][:30],marker = 'o',ls='--',color='g')
plt.plot(iris['petal_width'][:30],marker = 'o',ls='--',color='y')
plt.title("Sepal Length Analysis")
plt.ylabel("Sepal Length")
plt.xlabel("Sample Index")
plt.grid()
plt.legend(['Sepal Length','Sepal Width','Petal Length','Petal Width'])
plt.show()
```
