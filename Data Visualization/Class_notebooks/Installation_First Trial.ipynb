Typically, you import pyplot as 'plt' and use its functions to quickly generate and customize plots.

**3. Generate a simple plot using matplotlib.**

````python
import matplotlib.pyplot as plt
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
# Create a plot
plt.plot(x, y, marker='o')
# Show the plot
plt.show()
````

**4. Add a title and axis labels to the plot.**

````python
import matplotlib.pyplot as plt
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
# Add title and labels
plt.title('Sample Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# Create a plot
plt.plot(x, y, marker='o')
# Show the plot
plt.show()
````

**5. Change the marker style and line properties, add a legend to the plot, save the plot as an image file.**

````python
import matplotlib.pyplot as plt
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
# Change marker style
plt.plot(x, y, marker='x')  # Change marker to 'x'

# Change line style
plt.plot(x, y, linestyle='--', color='red')  # Change line style to dashed and color to red

#Add legend
plt.legend(['Data Points'], loc='upper left')

# Add title and labels
plt.title('Sample Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Create a plot
plt.plot(x, y, marker='o')
# # Show the plot
# plt.show()
plt.savefig('sample_plot.png')  # Save the plot as a PNG file
````
