```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import gaussian_kde
```

```python
url = 'https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv'
iris = pd.read_csv(url)
iris.head()
```

```python
# 1d kde
data = iris['petal_width']
kde_1 = gaussian_kde(data)
x = np.linspace(data.min(),data.max(),200)
plt.plot(x,kde_1(x),color='red')
plt.legend()
plt.grid()
plt.title('KDE of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Density')
plt.show()
```

```python
#2d kde and contour
x = iris['sepal_length']
y = iris['sepal_width']
xy = np.vstack([x,y])
kde_2d = gaussian_kde(xy)
xi, yi = np.mgrid[x.min():x.max():100j, y.min():y.max():100j]#
zi = kde_2d(np.vstack([xi.flatten(),yi.flatten()]))#
plt.figure(figsize=(8,6))#
plt.pcolormesh(xi,yi,zi.reshape(xi.shape), shading='auto', cmap='Blues') #
plt.contour(xi, yi, zi.reshape(xi.shape), colors='k')#
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('2D KDE and Contour Plot')
plt.colorbar(label='Density')
plt.show()
```
