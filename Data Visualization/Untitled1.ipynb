```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv'
iris = pd.read_csv(url)
iris.head()
```

```python
plt.scatter(iris['petal_length'],iris['petal_width'],color='green',alpha=0.3)
plt.title('Petal Length Vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
# error bars for any measurement error in any of the data
x = iris['petal_length'][:20]
y = iris['petal_width'][:20]
y_Err = np.random.uniform(0.05,0.2,size=20) #(low,high,size)
plt.errorbar(x,y,yerr=y_Err,fmt='o',color='blue',ecolor='red',capsize=4)
plt.title('Petal Length with error bars')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

```python
plt.hist(iris['sepal_length'],bins=20, color='brown',edgecolor='black',alpha=0.5)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
```
