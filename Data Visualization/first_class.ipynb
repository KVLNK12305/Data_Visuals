```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xp = [1,2,3,4,5]
yp = [2,4,6,8,10]
plt.plot(xp,yp,marker='x')
#titles for graph and axes
plt.title("sample graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```
