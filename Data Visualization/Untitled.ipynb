```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import gaussian_kde
```

```python
url = 'https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv'
iris = pd.read_csv(url)
fig, axs = plt.subplots(1, 2, figsize=(10, 10))

#plot1
sc1 = axs[0].scatter(iris['sepal_length'], iris['sepal_width'], c=iris['petal_length'], cmap='viridis')
plt.colorbar(sc1,ax=axs[0],label="Sepal Width")
axs[0].set_title('Sequential: Sepal Length vs Sepal Width')
axs[0].set_xlabel('Sepal Length (cm)')
axs[0].set_ylabel('Sepal Width (cm)')

sc2 = axs[1].scatter(iris['petal_length'],iris['petal_width'], c=iris["petal_length"],cmap="coolwarm")
plt.colorbar(sc2,ax=axs[1],label="Petal width")
axs[1].set_title('Sequential: Petal Length vs Petal Width')
axs[1].set_xlabel('Petal Length (cm)')
axs[1].set_ylabel('Petal Width (cm)')
plt.tight_layout()
plt.show()
```
